```python
import pandas as pd
from pandas import *
import json
```

```python
df = pd.read_csv('../static/globe-themes/globe-theme-data.csv')
```

```python
def format_number(x, decimals):
    a = float(str(x).rstrip('0'))
    a = round(a, decimals) if len(str(a).split('.')[1]) > decimals else a
    return a
```

```python
def compress(variable, filename):
    selection = df[['lat', 'lon', variable]]
    selection.columns = ['lat', 'lon', 'val']
    selection = selection.reset_index(drop=True)
    selection.lat = selection.lat.apply(lambda x: format_number(x, 3))
    selection.lon = selection.lon.apply(lambda x: format_number(x, 3))
    selection.val = selection.val.apply(lambda x: format_number(x, 1))
    selection.val = selection.val.apply(lambda x: '' if x == -32767 else str(x) if x > 0 else str(int(x)))
    
    tmp = {}

    for v in selection.values:
        if v[0] not in tmp:
            tmp[v[0]] = []

        tmp[v[0]].append(v[1])

        if (v[2] == ''):
            tmp[v[0]].append(-1)
        else:
            tmp[v[0]].append(float(v[2]))
            
    with open('../static/globe-themes/' + filename, 'w') as f:
        f.write(json.dumps(tmp, separators=(',', ':')))
```

```python
compress('HFO_s', 'too-much.json')
compress('DRO_s', 'too-little.json')
compress('ECO_S_s', 'too-dirty.json')
```
